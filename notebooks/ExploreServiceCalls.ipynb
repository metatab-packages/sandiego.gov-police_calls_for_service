```python
disp = pkg.resource('sdpd_calls_dispositions').dataframe()
beats = pkg.resource('sdpd_beats').dataframe()
call_types = pkg.resource('sdpd_call_type').dataframe()
ct_map = pkg.resource('ct_map').dataframe()
```

```python

calls = pkg.resource('sdpd_calls_for_service').dataframe()
calls.dtypes
```

```python
ct_map.head()
```

```python
df = calls.merge(beats, how='left')\
    .merge(disp.rename(columns={'description':'disp_desc'}), how='left')\
    .merge(call_types.rename(columns={'description':'call_type_desc'}), how='left')
df.head().T
```

```python
if False:
    df['is_arrest'] = df.description.str.contains('ARREST').astype(bool)
    df['is_transient'] = df.description.str.contains('TRANSIENT').astype(bool)
    df['is_mental'] = df.description.str.contains('MENTAL').astype(bool)
    df['is_domestic'] = df.description.str.contains('DOMESTIC').astype(bool)
    df['is_crisis'] = df.description.str.contains('CRISIS').astype(bool)
    df['is_toy'] = df.description.str.contains('TOY').astype(bool)
    df['is_vehicle'] = df.description.str.contains('VEHICLE').astype(bool)
```

```python
types = 'ARREST TRANSIENT VEHICLE MENTAL DOMESTIC CRISIS TOY '.split()
def simple_type(v):
    
    for t in types:
        if t in str(v):
            return t
    return 'OTHER'

df['simple_disp'] = df.disp_desc.apply(simple_type)
df.simple_disp.value_counts()
```
